## Combining and Merging Datsets

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
df1
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'd'], 'data2': range(3)})
df2
```

```python
pd.merge(df1,df2) #Merges the two together by key value
```

```python
pd.merge(df1, df2, on='key') #Speficy which column to merge on
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],'data2': range(3)})
df3
```

```python
df4
```

```python
pd.merge(df3,df4, left_on = "lkey", right_on = "rkey")  #speficy left and right columns
```

```python
pd.merge(df1, df2, how='outer') #Takes union of the keys to stop data from being cut
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],'data1': range(6)})
df1
```

```python
df2 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'], 'data2': range(5)})
df2
```

```python
pd.merge(df1, df2, on = "key", how="left") #how = left uses all keys in the left table (df1)
```

```python
pd.merge(df1, df2, how = "inner") #Creates n rows from data2 for each data1
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],'key2': ['one', 'two', 'one'], 'lval': [1, 2, 3]})
left
```

```python
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],'key2': ['one', 'one', 'one', 'two'],'rval': [4, 5, 6, 7]})
right
```

```python
pd.merge(left, right, on = ["key1", "key2"], how = "outer") #Merge through multiple columns
```

```python
 pd.merge(left, right, on='key1') #Merge together by index 
```

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right')) #Appends suffixes to overlapping column names
```

## Merging on Index

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],'value': range(6)})
left1
```

```python
right1 = pd.DataFrame({'group_val' : [3.5,7] }, index = ['a','b'])
right1
```

```python
pd.merge(left1, right1, left_on = "key", right_index = True) #use the index as the merge keys
```

```python
pd.merge(left1, right1, left_on = "key", right_index = True, how = "outer") #Form the union with all indexes
```

```python
lefth = pd.DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio',
 ....: 'Nevada', 'Nevada'],
 ....: 'key2': [2000, 2001, 2002, 2001, 2002],
 ....: 'data': np.arange(5.)})
lefth
```

```python
righth = pd.DataFrame(np.arange(12).reshape((6, 2)),
 ....: index=[['Nevada', 'Nevada', 'Ohio', 'Ohio',
 ....: 'Ohio', 'Ohio'],
 ....: [2001, 2000, 2000, 2000, 2001, 2002]],
 ....: columns=['event1', 'event2'])
righth
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True) #Merging hierarchically indexed data
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True, how='outer') #Use all keys
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
 ....: index=['a', 'c', 'e'],
 ....: columns=['Ohio', 'Nevada'])
left2
```

```python
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
 ....: index=['b', 'c', 'd', 'e'],
 ....: columns=['Missouri', 'Alabama'])
right2
```

```python
pd.merge(left2, right2, how = "outer", left_index = True, right_index = True) #Use index of both tables
```

```python
left2.join(right2, how='outer') #Merge with similar indexes
```

```python
left1.join(right1, on='key') #Join by column key
```

```python
another = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
 ....: index=['a', 'c', 'e', 'f'],
 ....: columns=['New York', 'Oregon'])
another
```

```python
left2.join([right2, another]) #Join mutiple DataFrames by the same key
```

```python
left2.join([right2, another], how='outer') #Merge by all keys
```

```python
arr = np.arange(12).reshape((3, 4))
arr
```

```python
np.concatenate([arr,arr], axis = 1) #Combines two numpy arrays
```

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])

pd.concat([s1,s2,s3]) #Combines the values and indexes down the columns
```

```python
pd.concat([s1,s2,s3]) #Concatenate down the rows
```

```python
pd.concat([s1, s2, s3], axis=1, sort = False) #Conctenate across the columns
```

```python
s4 = pd.concat([s1, s3])
s4
```

```python
pd.concat([s1,s4], axis = 1, sort = False)
```

```python
pd.concat([s1, s4], axis=1, join='inner') #Only uses keys from both
```

```python
pd.concat([s1, s4], axis=1, sort = False).reindex(['a', 'c', 'b', 'e']) #Specify new index labels
```

```python
#Concetenate and create heiarchical index to distinguish
result = pd.concat([s1, s1, s3], keys=['one', 'two', 'three']) 
result
```

```python
result.unstack()
```

```python
pd.concat([s1, s2, s3], axis=1, keys=['one', 'two', 'three'], sort = False) #Keys become column headers
```

## Combining Data with Overlap


```python
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan], index=['f', 'e', 'd', 'c', 'b', 'a'])
b = pd.Series(np.arange(len(a), dtype=np.float64), index=['f', 'e', 'd', 'c', 'b', 'a'])
```

```python
b[-1] = np.nan
```

```python
a
```

```python
b
```

```python
np.where(pd.isnull(a), b, a ) #Where a is null, replace it with b
```

```python
b[:-2].combine_first(a[2:]) #Same operation of patching values
```

```python
df1 = pd.DataFrame({'a': [1., np.nan, 5., np.nan], 'b': [np.nan, 2., np.nan, 6.], 'c': range(2, 18, 4)})
df2 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],'b': [np.nan, 3., 4., 6., 8.]})
```

```python
df1
```

```python
df2
```

```python
df1.combine_first(df2) #Patch NaN values from df1 with df2
```

## Reshaping and Pivoting

```python
data = pd.DataFrame(np.arange(6).reshape((2, 3)),
 .....: index=pd.Index(['Ohio', 'Colorado'], name='state'),
 .....: columns=pd.Index(['one', 'two', 'three'],
 .....: name='number'))

data
```

```python
data.stack() #Pirvots the columns into the rows, producing a series
```

```python
result = data.stack()
result.unstack() #Pivots the wors back into columns to produce a DataFrame
```

```python
result.unstack(0) #Pivot the columns to rows
```

```python
result.unstack('state') #Pivots the states from the rows to the columns
```

```python
s1 = pd.Series([0, 1, 2, 3], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([4, 5, 6], index=['c', 'd', 'e'])
s1
```

```python
s2
```

```python
data2 = pd.concat([s1, s2], keys=['one', 'two'])
data2
```

```python
data2.unstack() #Produces NaN if not found
```

```python
data2.unstack().stack()
```

```python
data2.unstack().stack(dropna=False) #Keep NaN values
```
